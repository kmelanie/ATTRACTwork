```python
###collect data using different samplepathlists generated above 
### create dictionary data consisting of dictionarys with sampleID as key and 12 datatype-dictionarys as values
data={}
def collectdata(samplepathlist, typeofdata):
    global data
    if not data:
        for i in samplepathlist:
            data[i[1]]=[]
    for i in samplepathlist:   
        dictname=typeofdata+"dict"
        dictname={typeofdata:[]}
        with open("../benchmark5/"+i[0]) as datafile:
            for line in datafile:
                dictname[typeofdata].append(line.rstrip().split(" ")[-1])

        datafile.close()
        #print(data[i[1]])
        data[i[1]].append(dictname)
    #print(data)
    
# ##rigid data
# collectdata(ligandr_energy_rigid, 'rigid_energy_score')
# collectdata(ligandr_refe_energy_rigid, 'rigid_energy_score_refe')

# collectdata(intra_irmsd_rigid,'rigid_intra_iRMSD')
# collectdata(intra_refe_irmsd_rigid, 'rigid_intra_refe_iRMSD')
# collectdata(inter_irmsd_rigid, 'rigid_inter_iRMSD')

# ## flexible data
# collectdata(intra_rmsca, 'intra_rmsca')
# collectdata(intra_refe_rmsca, 'intra_refe_rmsca')
# collectdata(inter_rmsca, 'inter_rmsca')

collectdata(ligandr_energy, 'energy_score')
collectdata(ligandr_refe_energy, 'energy_score_refe')

collectdata(ligandr_energy_restraint, 'restraining_energy_score')
collectdata(ligandr_refe_energy_restraint, 'restraining_energy_score_refe')

collectdata(intra_irmsd,'intra_iRMSD')
collectdata(intra_refe_irmsd, 'intra_refe_iRMSD')
collectdata(inter_irmsd, 'inter_iRMSD')

# collectdata(ligonly_intra_rmsca, 'ligonly_intra_rmsca')
# collectdata(ligonly_intra_refe_rmsca, 'ligonly_intra_refe_rmsca')
# collectdata(ligonly_inter_rmsca, 'ligonly_inter_rmsca')

# collectdata(looponly_intra_rmsca, 'looponly_intra_rmsca')
# collectdata(looponly_intra_refe_rmsca, 'looponly_intra_refe_rmsca')
# collectdata(looponly_inter_rmsca, 'looponly_inter_rmsca')



#print(data)
        
```

```python
table=[['PDB', 'energy_score_start', 'energy_score_end', 'energy_score_refe_start', 'energy_score_refe_end','restraining_energy_score_start', 'restraining_energy_score_end','restraining_energy_score_refe_start', 'restraining_energy_score_refe_end' ,'intra_iRMSD','intra_refe_iRMSD', 'inter_iRMSD_start','inter_iRMSD_end']]
for sampleID in data:
    #print(sampleID)
    sampleID_list=[]
    for measurement in data[sampleID]:
        #print(measurement)
        for classification in measurement:
            #print('start', sampleID, classification, measurement[classification][0])
            #print('end', sampleID, classification, measurement[classification][-1])
            if not sampleID_list:
                sampleID_list.append(sampleID)
            if float(measurement[classification][0])==0.000:
                sampleID_list.append(measurement[classification][-1])
            else:
                sampleID_list.append(measurement[classification][0]+'\t'+measurement[classification][-1])
                
    table.append(sampleID_list)
            
#print(sampleID_list)
#print(table)
with open('all_data_sep.txt', 'w') as output: 
    for i in table:
        #print(i)
        print('\t'.join(map(str, i)), file=output)
        
```
